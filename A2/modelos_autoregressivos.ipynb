```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -----------------------------
# 1. Leitura e preparação
# -----------------------------
df = pd.read_csv("data_updated.csv")
df["week"] = pd.to_datetime(df["week"])
df = df.set_index("week")
y = df["volume"]

# -----------------------------
# 2. Funções de erro
# -----------------------------
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# -----------------------------
# 3. Função geral para treinar, prever e plotar
# -----------------------------
def fit_and_evaluate(model_name, order=(1,0,0), seasonal_order=None, h=4):
    
    # Separar treino e teste
    train = y[:-h]
    test = y[-h:]
    
    # Ajustar modelo
    if seasonal_order is None:
        model = ARIMA(train, order=order)
    else:
        model = SARIMAX(train, order=order, seasonal_order=seasonal_order)
    
    fit = model.fit()
    
    # Fitted values e forecast
    fitted = fit.fittedvalues
    fitted.index = train.index[-len(fitted):]
    
    pred = fit.get_forecast(steps=h)
    pred_mean = pred.predicted_mean
    pred_ci = pred.conf_int()
    pred_mean.index = test.index
    pred_ci.index = test.index
    
    # Plot
    plt.figure(figsize=(12,5))
    plt.plot(y.index, y, label="Observado", linewidth=2)
    plt.plot(fitted.index, fitted, label=f"Ajustado — {model_name}", color="green")
    plt.plot(pred_mean.index, pred_mean, label=f"Forecast — {model_name}", color="red")
    plt.fill_between(pred_ci.index, pred_ci.iloc[:,0], pred_ci.iloc[:,1], color="red", alpha=0.25)
    plt.title(f"{model_name} — Observado, Ajustado e Forecast")
    plt.xlabel("Week")
    plt.ylabel("Volume")
    plt.grid(True)
    plt.legend()
    plt.show()
    
    # Calcular métricas
    results = {
        "Modelo": model_name,
        "RMSE": rmse(test, pred_mean),
        "MAE": mae(test, pred_mean),
        "MAPE (%)": mape(test, pred_mean)
    }
    
    return results

# -----------------------------
# 4. Lista de modelos
# -----------------------------
models = [
    ("AR(1)", (1,0,0), None),
    ("AR(2)", (2,0,0), None),
    ("AR(3)", (3,0,0), None),
    ("MA(1)", (0,0,1), None),
    ("MA(2)", (0,0,2), None),
    ("ARMA(1,1)", (1,0,1), None),
    ("SARIMA(1,0,0)(1,0,0)[52]", (1,0,0), (1,0,0,52)),
    ("SARIMA(1,0,1)(1,0,0)[52]", (1,0,1), (1,0,0,52))
]



# -----------------------------
# 5. Rodar modelos e coletar resultados
# -----------------------------
h = 4  # Número de passos a prever
results = []



for name, order, seas in models:
    print(f"\n===== MODELO: {name} =====")
    res = fit_and_evaluate(name, order, seas, h=h)
    results.append(res) 
    
    # Plot ACF e PACF dos resíduos
    fit = ARIMA(y, order=order).fit() if seas is None else SARIMAX(y, order=order, seasonal_order=seas).fit()
    residuals = fit.resid

    plt.figure(figsize=(10,4))
    plot_acf(residuals, lags=20, alpha=0.05)
    plt.title(f"ACF dos Resíduos — {name}")
    plt.show()

    plt.figure(figsize=(10,4))
    plot_pacf(residuals, lags=20, alpha=0.05, method='ywm')
    plt.title(f"PACF dos Resíduos — {name}")
    plt.show()


# -----------------------------
# 6. Tabela final comparativa
# -----------------------------
df_results = pd.DataFrame(results).sort_values(by="RMSE")
print("\n===== COMPARAÇÃO FINAL DOS MODELOS =====\n")
print(df_results)
```
